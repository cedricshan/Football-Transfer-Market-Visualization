```python
folder_path = 'data'
processed_folder = os.path.join(folder_path, 'processed_data')

# Check if the 'processed_data' folder exists, if not, create it
if not os.path.exists(processed_folder):
    os.makedirs(processed_folder)

for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path):
        df=pd.read_csv(file_path)
        to_drop=[
                'Club',
                'Player',
                'Pos',
                'Club Involved',
                'Transfer Movement',
                'League Name'
            ]
        df.drop(to_drop, inplace=True, axis=1)
        df.dropna(inplace=True)
        filtered_df = df[df['Fee'].str.contains('€', na=False)].copy()
        filtered_df['Fee'] = filtered_df['Fee'].str.replace('€', '')
        filtered_df['Fee'] = filtered_df['Fee'].drop(filtered_df[filtered_df['Fee'].str.contains('Loan fee', na=False)].index)

        mask_m = filtered_df['Fee'].str.contains('m', na=False)
        filtered_df.loc[mask_m, 'Fee'] = (filtered_df.loc[mask_m, 'Fee']
                                        .str.replace('m', '')
                                        .astype(float) * 1000000)

        mask_k = filtered_df['Fee'].str.contains('k', na=False)
        filtered_df.loc[mask_k, 'Fee'] = (filtered_df.loc[mask_k, 'Fee']
                                        .str.replace('k', '')
                                        .astype(float) * 1000)

        filtered_df['Fee'] = filtered_df['Fee'].astype(float)
        filtered_df.dropna(inplace=True)
        filtered_df.to_csv(f'processed_{filename}')
        
```
